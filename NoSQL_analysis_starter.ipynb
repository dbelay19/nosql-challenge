### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
score_20 = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
score_count = establishments.count_documents(score_20)
print(f"There are {score_count} establishments with a hygiene score of 20.")
#count = establishments.count_documents(score_20)
# Display the first document in the results using pprint
pprint(establishments.find_one(score_20))
```

```python
# Convert the result to a Pandas DataFrame
hygiene_score_df = pd.DataFrame(establishments.find(score_20))

# Display the number of rows in the DataFrame
print(f"There are {len(hygiene_score_df)} rows in this DataFrame.")

# Display the first 10 rows of the DataFrame
hygiene_score_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
rating_query = {'LocalAuthorityName': {'$regex': "London"}, 'RatingValue': {'$gte': '4'}}
london_count = establishments.count_documents(rating_query)
print(f"There are {london_count} establishments in London that have a RatingValue greater than or equal to 4.")

# Display the first document in the results using pprint
london_count = establishments.find_one(rating_query)
if london_count:
    pprint(london_count)
else:
    print("No establishments found that meet the criteria.")
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(establishments.find(rating_query))

# Display the number of rows in the DataFrame
#print(f"There are {len(london_df)} rows in this DataFrame.")

print(f"There nuber of rows in this DataFrame is {len(london_df)}.")

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
score_query = {'$match': {'scores.Hygiene': 0}}
authority_query = {'$group': {'_id': '$LocalAuthorityName', 
                          'count': {'$sum': 1}
                         }
              }
sort = {'$sort': {'count': -1}}

pipeline = [score_query , authority_query, sort]

query_results = list(establishments.aggregate(pipeline))
# Print the number of documents in the result
print(f"There are {len(results)} documents in the result.")
# Print the first 10 results
pprint(query_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
aggregated_df = pd.json_normalize(query_results)
# Display the number of rows in the DataFrame
print(f"There number of rows in the DataFrame are {len(aggregated_df )}.")
# Display the first 10 rows of the DataFrame
aggregated_df.head(10)
```
